```python
# Importing the necessary library
import requests
import os
import json
from google.colab import files
```

```python
url_base = "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=f58a46bc624beb6ac65ab32a1e4a800d&file_type=json"

file_base = "data_FRED_{}.json"
fredSeries = ['NYGDPPCAPKDJPN','MSPUS','SLUEM1524ZSROU','LNS14000006']

# Loop
for i in fredSeries:  

   print(i) 
   URL = url_base.format(i)
   print(URL)

   data = requests.get(URL).json()
   print(data)

   fileName = file_base.format(i)
   print(fileName)

   # Downloading
   with open(fileName, 'w', encoding='utf-8') as f:
     json.dump(data, f, ensure_ascii=False, indent=4)

   files.download(fileName)
```
