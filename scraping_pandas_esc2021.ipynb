# Scraping Wikipedia Eurovision 2021 data

using `pandas`



```python
import requests, pandas as pd
```

```python
tables=pd.read_html('https://en.wikipedia.org/wiki/Eurovision_Song_Contest_2021')
```

```python
tables[6]
```

```python
df=tables[6]
```

```python
df.columns
```

```python
df.columns=['draw','country','artist','song', 'language','place','points']
```

```python
df
```

```python
df.sort_values(by='points',ascending=False)
```

```python
df=df.sort_values(by='points',ascending=False)
```

Cleaning the columns I need for the chart

```python
df['place'].str.strip('[q]').astype(int)
```

```python
df['place']=df['place'].str.strip('[q]').astype(int)
```

```python
df
```

```python
df['country'].str.strip('[o]')
df['country']=df['country'].str.strip('[o]')
```

```python
df
```

```python
df.to_csv('ESC2021_results.csv')
```
